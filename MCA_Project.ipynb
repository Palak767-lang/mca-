```python
# ð Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ð Step 2: Create Dataset (Temperature vs Revenue)
data = {
    "Temperature": [20, 22, 25, 27, 30, 32, 35, 37, 40],
    "Revenue":     [200, 220, 250, 270, 300, 320, 350, 370, 400]
}
df = pd.DataFrame(data)
print("Sample Data:")
print(df)

# ð Step 3: Visualize Data
plt.scatter(df["Temperature"], df["Revenue"], color="blue")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice Cream Revenue (â¹)")
plt.title("Temperature vs Revenue")
plt.show()

# ð Step 4: Prepare Data for Model
X = df[["Temperature"]]   # Features
y = df["Revenue"]         # Target

# ð Step 5: Train Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# ð Step 6: Predictions
y_pred = model.predict(X)

# ð Step 7: Evaluate Model
print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("RÂ² Score:", r2_score(y, y_pred))

# ð Step 8: Plot Regression Line
plt.scatter(X, y, color="blue", label="Actual")
plt.plot(X, y_pred, color="red", label="Predicted")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice Cream Revenue (â¹)")
plt.title("Ice Cream Revenue Prediction")
plt.legend()
plt.show()

# ð Step 9: Test with New Temperature
new_temp = pd.DataFrame({"Temperature": [28]}) # Convert to DataFrame with feature name
predicted_revenue = model.predict(new_temp)
print("Predicted Revenue for 28Â°C:", predicted_revenue[0])
```

```python
# Calculate Residuals
residuals = y - y_pred

# Visualize Residuals
plt.scatter(y_pred, residuals, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Revenue (â¹)")
plt.ylabel("Residuals (Actual - Predicted Revenue) (â¹)")
plt.title("Residual Plot")
plt.show()
```
